```python
import json
import copy
```

```python
old_json_path = '/home/adichoudhary/CSE523/vehicle-ReId/repo/data/dataset_1.json'
```

```python
with open(old_json_path, "r") as read_file:
    data = json.load(read_file)
```

```python
data.keys()
```

```python
data_all = copy.deepcopy(data)
```

```python
data['Set04']
```

```python
surr = 'classes_surr'
surr_all = 'classes_surrall'
```

```python
for num in range(1,6):
    
    set_num = "Set0"+str(num)
    
    for row in data[set_num]:
        row[0][0] = row[0][0].replace('classes',surr)
        row[2][0] = row[2][0].replace('classes',surr)
        
        for i in range(4):
            row[i][0] = row[i][0].replace('data/','data_new/')
            row[i][0] = row[i][0].replace('dataset2','dataset')
```

```python
data['Set03']
```

```python
for num in range(1,6):
    
    set_num = "Set0"+str(num)
    
    for row in data_all[set_num]:
        row[0][0] = row[0][0].replace('classes',surr_all)
        row[2][0] = row[2][0].replace('classes',surr_all)
        
        for i in range(4):
            row[i][0] = row[i][0].replace('data/','data_new/')
            row[i][0] = row[i][0].replace('dataset2','dataset')
```

```python
file_s = '/home/adichoudhary/CSE523/vehicle-ReId/repo/data_new/dataset_surr.json'
file_s_all = '/home/adichoudhary/CSE523/vehicle-ReId/repo/data_new/dataset_surrall.json'
```
